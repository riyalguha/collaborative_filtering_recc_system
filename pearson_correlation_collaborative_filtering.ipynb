```python
places = geodata[['placeID','name']]
```

```python
places.head()
```

```python
rating = pd.DataFrame(frame.groupby('placeID')['rating'].mean())
```

```python
rating.head()
```

```python
rating['rating_count'] = pd.DataFrame(frame.groupby('placeID')['rating'].count())
```

```python
rating.head()
```

```python
rating.describe()
```

```python
rating.sort_values('rating_count',ascending = False)
```

```python
places[places['placeID'] == 135085]
```

```python
cuisine[cuisine['placeID'] == 135085]
```

```python
places_crosstab = pd.pivot_table(data = frame,values='rating',index = 'userID', columns = 'placeID')
```

```python
places_crosstab.head()
```

```python
Tortas_rating = places_crosstab[135085]
```

```python
Tortas_rating[Tortas_rating >= 0]
```

```python
similar_tortas = places_crosstab.corrwith(Tortas_rating)
```

```python
corr_tortas = pd.DataFrame(similar_tortas , columns = ['PearsonR'])
```

```python
corr_tortas.head()
```

```python
corr_tortas.dropna(inplace = True)
corr_tortas.head()
```

```python
corr_tortas.sort_values('PearsonR',ascending = False).head()
```

```python
Tortas_corr_summary = corr_tortas.join(rating['rating_count'])
```

```python
Tortas_corr_summary[Tortas_corr_summary['rating_count'] > 8].sort_values('PearsonR', ascending = False).head(10)
```
